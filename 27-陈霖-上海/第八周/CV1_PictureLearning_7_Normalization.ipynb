```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Normalization(x):
    return [(float(i)-np.mean(x))/(max(x)-min(x)) for i in x]
```

```python
def z_score(x):
    x_mean=np.mean(x)
    s2=sum([(i-np.mean(x))*(i-np.mean(x)) for i in x])/len(x)
    s3=np.std(x)
    print(s2,s3*s3)
    return [(i-x_mean)/s3 for i in x]
```
